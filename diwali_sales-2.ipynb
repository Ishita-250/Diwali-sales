Gender

```python
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python

fig, axes=plt.subplots(1,2,figsize=(12,5))
a=sns.countplot(x="Gender",data=data,palette="dark",ax=axes[0])
axes[0].set_title("GENDER COUNT")
for bars in a.containers:
    a.bar_label(bars)


axes[1].bar(sales_gen["Gender"],sales_gen["Amount"],color="salmon")
axes[1].set_title("SALES BY GENDER")
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("total sales amount")
```

From above graph we can see that most of the buyers are females and the most amount spent is also done by females.

Age

```python
fig, axes=plt.subplots(1,2,figsize=(12,5))
a=sns.countplot(data=data,x="Age Group",hue ="Gender",ax=axes[0])
for bars in a.containers:
    a.bar_label(bars)

sales_age=data.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
axes[1].bar(sales_age["Age Group"],sales_age["Amount"])
axes[1].set_xlabel("Age Group")
axes[1].set_ylabel("Amount")
axes[1].set_title("Amount spent by different age group")
axes[0].set_title("number of males and females")
```

From above graph we can see that most of the purchases are in age group of 26-35 years by females.

State

```python
sales_state = data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
fig, axes=plt.subplots(1,2,figsize=(12,5))
a=sns.barplot(data = sales_state, x = 'State',y= 'Orders',ax=axes[0])
axes[0].tick_params(axis='x', labelrotation=90) 
for bars in a.containers:
    a.bar_label(bars)
axes[0].set_title("number of orders in each state")


sales_am = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
axes[1].set_title("Amount spent by each state")
axes[1].set_xlabel("State")
axes[1].set_ylabel("Amount")
axes[1].bar(sales_am["State"],sales_am["Amount"])
axes[1].tick_params(axis='x', labelrotation=90) 

plt.tight_layout()
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka.

Occupation

```python
fig, axes=plt.subplots(2,1,figsize=(20,15))
a = sns.countplot(data = data, x = 'Occupation',palette="coolwarm",ax=axes[0])
for bars in a.containers:
    a.bar_label(bars)
axes[0].set_title("Number of orders in each occupation",size=24)

color=sns.color_palette("pastel")
sales_occ=data.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
a=axes[1].bar(sales_occ["Occupation"],sales_occ["Amount"],color=color)
axes[1].set_title("Amount spent by each occupation",size=24)
axes[1].set_xlabel("Occupation")
axes[1].set_ylabel("Amount")
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

Martial Status

```python
fig, axes=plt.subplots(1,2,figsize=(12,7))
a=sns.countplot(data=data,x="Marital_Status",ax=axes[0],palette="Set3",hue="Gender")
for bars in a.containers:
    a.bar_label(bars)
axes[0].set_title("Count of orders done by single and married people")

sales_mar=data.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
b=sns.barplot(data=sales_mar,x="Marital_Status",y="Amount",ax=axes[1],palette="Set3",hue="Gender")
axes[1].set_title("Amount sent by prople based on their marital status")

```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

Product Category

```python
fig ,axes=plt.subplots(2,1,figsize=(20,19))
a=sns.countplot(data=data,x="Product_Category",ax=axes[0],palette="muted")
axes[0].tick_params(axis='x', labelrotation=60) 
for bars in a.containers:
    a.bar_label(bars)
axes[0].set_title("Count of products sold",size=23)

axes[1].set_title("Amount spent on each product category",size=23)
sales_prod=data.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount")
b=sns.barplot(data=sales_prod,x="Product_Category",y="Amount",ax=axes[1],palette="coolwarm")
axes[1].tick_params(axis= 'x', labelrotation= 60)
plt.tight_layout()
```

Here, we can see that the most ordered products categories purchased are cleaning & apparel, food and electronic.
Also the product category on which most of the amount is spent is food.
